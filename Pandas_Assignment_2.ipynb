```python
#Return a count of how many times each car manufacturer appears in the data. Order your results in descending order.
df['car_make'].value_counts(ascending=False)
```

```python
# Return a count of how many cars are in each state. Order your results in descending order.
df['state'].value_counts(ascending=False)
```

```python
#Filter the dataframe to show only cars made by Lexus where the model is ES
new_df= df.loc[(df.car_make == 'Lexus') & (df.car_model == 'ES'), :]
new_df
```

```python
# Filter the dataframe to show only fords made after the year 2000 in the state of California
Ford_df= df.loc[(df.car_make == 'Ford') & (df.car_model_year > 2000) & (df.state == 'California'), :]
Ford_df
```

```python
#Filter the dataframe to show Audi's or Porsche's in the state of Texas where the model year is before the year 2000
AudiPorsche_df= df.loc[(df.car_make.isin(['Audi', 'Porsche'])) & (df.car_model_year < 2000) & (df.state == 'Texas'),:]
AudiPorsche_df
```

```python
#Create a column named "classic car". If the model year is 2001 or before then put  a "yes"  in the column, for all other cars put a "no" in this column.

result = []
for value in df["car_model_year"]:
    if value <= 2001:
        result.append("Yes")
    else:
        result.append("No")
       
df["classic_car"] = result  
print(df)
```
