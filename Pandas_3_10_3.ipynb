```python
df = pd.read_csv('Salaries.csv')
```

```python
df.head()
```

```python
#Create a column named 'EmployeeFirstName' and extract the first name from the EmployeeName Column
df['EmployeeFirstName'] = df.EmployeeName.str.split().str.get(-3)
df
```

```python
#Create a column named 'EmployeeLasttName' and extract the first name from the EmployeeName Column
df['EmployeeLastName'] = df.EmployeeName.str.split().str.get(-1)
df
```

```python
#Convert all letters in the JobTitle column to all upper case
df['JobTitle'].str.upper()
```

```python
#Find the average TotalPay of a transit operator
df[df['JobTitle'].str.contains('Transit Operator') & df['TotalPay']].mean()
df['TotalPay'].mean()
```

```python
# Find the top 10 employees with the highest Total pay
df.sort_values(['TotalPay'], ascending=False).groupby('TotalPay').head(10)
df.TotalPay.head(10)
```

```python
#Find all employee where there OtherPay exceeds $100,000
df[df['OtherPay'] > 100000]
```

```python
#Find the mean TotalPay where the JobTitle contains the word "Police"
df[df['JobTitle'].str.contains('Police') & df['TotalPay']].mean()
df['TotalPay'].mean()
```

```python
#Find the  person with the Max TotalPay where the JobTitle contains the word "Police"
df[df['JobTitle'].str.contains('Police')].head(1)
```

```python
#Find the highest paid nurse.
df[df['JobTitle'].str.contains('Nurse')].head(1)
```
