```python
def CalculateTaxes(income):
  Deduction = 12000
  global TotalTax
  TotalTax = 0
  if income <= Deduction:
    TotalTax = 0
  else:
      income= income-Deduction
      rate = .10
      if income <=9950:
        tax = income * rate
      else:
        tax = 9950 * rate

#print(str(tax))
      TotalTax = tax +TotalTax
      SecondBracket(income)
      ThirdBracket(income)
      FourthBracket(income)
      FifthBracket(income)
      SixthBracket(income)
      SeventhBracket(income)
      return "${:,.2f}" .format(TotalTax)
    
##############################################################################################################################################

def SecondBracket(income):
  global TotalTax
  rate = .12
  previouslimit = 9950
  newlimit = 40525
  
  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax

#################################################################################################################################################

def ThirdBracket(income):
  global TotalTax
  rate = .22
  previouslimit = 40525
  newlimit = 86375
  
  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax

#################################################################################################################################################

def FourthBracket(income):
  global TotalTax
  rate = .24
  previouslimit = 86375
  newlimit = 164925

  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax
    
###############################################################################################################################################

def FifthBracket(income):
  global TotalTax
  rate = .32
  previouslimit = 164925
  newlimit = 209425
  
  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax

###################################################################################################

def SixthBracket(income):
  global TotalTax
  rate = .32
  previouslimit = 209425
  newlimit = 523600
  
  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax

#####################################################################################################

def SeventhBracket(income):
  global TotalTax
  rate = .32
  previouslimit = 523600
  newlimit = income
  
  if income <= previouslimit:
    tax = 0
  else:
    if income <= newlimit:
        tax = (income - previouslimit) * rate
    else:
          tax = (newlimit - previouslimit) * rate  
          TotalTax = tax + TotalTax
###################################################################################################

TotalTax = 0
```
