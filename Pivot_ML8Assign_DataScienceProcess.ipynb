### As we did earlier in the week, we will be interested in using the Brand of the refrigerators as a Feature and/or a Target. As such, we will need to convert the Brand to a number as we did with AddressCity and CityNum.

```python
# 5
# Let's find which brands (and how many of each) are represented:

Refrigerators['Brand'].value_counts() # Again, you may take a look at ML8 for a reference.
```

```python
# Define the new column and set initial values = 0
Refrigerators['BrandNum'] = 0

# Create a loop to assign the additional numbers to the remaining Brands.
for i in range(1,len(Refrigerators['Brand'].unique())):
  Refrigerators.loc[Refrigerators['Brand'] == Refrigerators['Brand'].unique()[i],'BrandNum']=i
  
# View a sample of 10 rows of the 'Brand' and 'BrandNum' of Refrigerators.
Refrigerators[['Brand','BrandNum']].sample(10)
```

```python
# We might also be interested in how many of each particular Style are represented:

Refrigerators['Style'].value_counts()
```

'# 6'
### Modify the code below to convert the refrigerator Style to a number ... StyleNum, so that it can be used as a Feature and Target as we proceed through the notebook.

```python
# 7
# Again ... change the code below to reflect Style instead of Brand!!!!!
# Define the new column and set initial values = 0
Refrigerators['StyleNum'] = 0

# Create a loop to assign the additional numbers to the remaining Styles.
for i in range(1,len(Refrigerators['Style'].unique())):
  Refrigerators.loc[Refrigerators['Style'] == Refrigerators['Style'].unique()[i],'StyleNum']=i
  
# View a sample of 10 rows of the 'Style' and 'StyleNum' of Refrigerators.
Refrigerators[['Style','StyleNum']].sample(10)
```

## Create a Model

### 1) Choose the Features and a Target for your problem.


```python
Refrigerators.columns
```

```python
# Define the features and the target from our dataset.

# Features
X = Refrigerators[['BrandNum','Width','Height','Depth','Capacity','StyleNum']]

# Target
y = Refrigerators['Price']
```

### 2) Create Training and Testing sets

```python
# import the package to create the Training and Testing split.
from sklearn.model_selection import train_test_split

# The code below splits our X and y into training and testing sets. 
# Since test_size=0.30, the split will be 70% Training to 30% Testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2)
```

'# 8'
### The Train / Test split is 70% for the Training Set and 30% for the Test Set. 

### So, approximately how many data points are in each set?

If there is 400 referigerators, 70% is 315 for the training set and 30% is 135 for the test set.

### 3) Feature Scaling

```python
# We now transform our feature values to the number
# of Standard Deviations each value is from its mean.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X_train[:5]
```

## Multiple Linear Regression: Select, Instantiate, Fit, Score

```python
## Select/Import the Model.
from sklearn import linear_model
    
## Instantiate the Linear Regression model. 
MLR = linear_model.LinearRegression()

## Fit the model on the training data.
MLR.fit(X_train, y_train)
    
## See how the model performs on the test data.
MLR.score(X_test, y_test)
```

### Multiple Linear Regression: Evaluate ... ScatterPlots

```python
# Predicted values of y based upon the model
y_pred = MLR.predict(X_test)

from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

fig = pyplot.figure()
ax = Axes3D(fig)

ax.scatter(X_test[:,0],X_test[:,1], y_test,  color='black')
ax.scatter(X_test[:,0],X_test[:,1], y_pred,  color='blue')
pyplot.show()
```

## K Nearest Neighbors: Select, Instantiate, Fit, Score

```python
## Select/Import the Model.
from sklearn.neighbors import KNeighborsRegressor
    
## Instantiate the Knn model with ... neighbors. 
KNN = KNeighborsRegressor(n_neighbors=10)
    
## Fit the model on the training data.
KNN.fit(X_train, y_train)
    
## See how the model performs on the test data.
KNN.score(X_test, y_test)
```

##Random Forest: Select, Instantiate, Fit, Score

```python
## Select/Import the Model.
from sklearn.ensemble import RandomForestRegressor

## Instantiate the Random Forest model
RF = RandomForestRegressor(n_estimators=1)

## Fit the model on the training data.

RF.fit(X_train, y_train)

## See how the model performs on the test data.
print(RF.score(X_test, y_test))
```

### Let's move on to consider the **Classification** models. We will again use the features 'Width', 'Height', 'Depth', 'Capacity', and 'BrandNum', as well as 'Price', but this time our target will be 'StyleNum'.

---



### 1) Choose the Features and a Target for your problem.

```python
# Define the features and the target from our dataset.
# Note that based upon the quantitive data that is given,
# We will try to predict the 'BrandNum' of a refrigerator.

# Features
X = Refrigerators[['Width', 'Height', 'Depth', 'Capacity', 'BrandNum', 'Price']]

# Target
y = Refrigerators['StyleNum']
```

### 2) Create Training and Testing sets

```python
# import the package to create the Training and Testing split.
from sklearn.model_selection import train_test_split

# The code below splits our X and y into training and testing sets. 
# Since test_size=0.30, the split will be 70% Training to 30% Testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2)
```

### 3) Feature Scaling

```python
# We now transform our feature values to the number
# of Standard Deviations each value is from its mean.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X_train[:5]
```

## K Nearest Neighbors: Select, Instantiate, Fit, Score


```python
## Select/Import the Model.
from sklearn.neighbors import KNeighborsClassifier
    
## Instantiate the Knn model with ... neighbors. 
KNN = KNeighborsClassifier(n_neighbors=2)
    
## Fit the model on the training data.
KNN.fit(X_train, y_train)
    
## See how the model performs on the test data.
KNN.score(X_test, y_test)
```

'# 11'
### Take the time to change the number of neighbors in the KNN model above to see if you can increase the score.

###K Nearest Neighbors: Evaluate ... Classification Report

```python
# Predicted values of y based upon the model
y_pred = KNN.predict(X_test)

# The classification report shows how well the model performed on each possible outcome
# or CityNum in our case.
from sklearn.metrics import classification_report  

# The classification report compares what was predicted for the test set  
# against what was actually seen in the test set.
print(classification_report(y_test, y_pred)) 
```

##Random Forest: Select, Instantiate, Fit, Score

```python
## Select/Import the Model.
from sklearn.ensemble import RandomForestClassifier

## Instantiate the Random Forest model
RF = RandomForestClassifier(n_estimators=30)

## Fit the model on the training data.
RF.fit(X_train, y_train)

## See how the model performs on the test data.
print(RF.score(X_test, y_test))
```

'# 12'
### Again, take some time to change the number of estimators in the Random Forest model above. See if you can increase the prediction score.

### Random Forest: Evaluate ... Classification Report




```python
# Predicted values of y based upon the model
y_pred = RF.predict(X_test)

# The classification report shows how well the model performed on each possible outcome
# or CityNum in our case.
from sklearn.metrics import classification_report  

# The classification report compares what was predicted for the test set  
# against what was actually seen in the test set.
print(classification_report(y_test, y_pred)) 
```
